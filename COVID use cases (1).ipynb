# Corona Virus Live Tracking For INDIA

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df
```

```python
max(df['new_death'])
```

```python
df = df.iloc[:-1,1:]
df.tail()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
total_states = np.arange(len(df['state_name']))
total_states
```

```python
df['active']
```

```python
max(df['positive'])
```

## Total Positive Cases Based On States in India

```python
from matplotlib.pyplot import figure
figure(num = None, figsize = (9,6), dpi = 80, facecolor = 'w', edgecolor = 'k')
plt.barh(total_states,df['positive'],align = 'center',alpha = 0.5, color = (1,0,0), edgecolor = (0.5,0.2,0.8))
plt.yticks(total_states, df['state_name'])
plt.xlim(1,max(df['positive']) +100)
plt.xlabel('positive Number of Cases')
plt.title("corona virus POSITIVE cases")
plt.show()
```

## Active Number of Cases Based on States

```python
from matplotlib.pyplot import figure
figure(num = None,figsize = (9,6), dpi = 80, facecolor = 'cyan', edgecolor = 'k')
plt.barh(total_states,df['new_active'], align = 'center', alpha = 0.5, color = (1,0,0), edgecolor = (1,1,1))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['new_active']) + 10)
plt.xlabel('Active Number of  Cases')
plt.title('Corona Virus NEW ACTIVE Cases')
plt.show()
```

## Death Cases Based on States

```python
from matplotlib.pyplot import figure
figure(num = None,figsize = (9,6), dpi = 80, facecolor = 'bisque', edgecolor = 'k')
plt.barh(total_states,df['death'], align = 'center', alpha = 0.5, color = 'maroon', edgecolor = (0.5,0.4,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death']) +10)
plt.xlabel('Death Number of Cases')
plt.title('Corona Virus DEATH Cases')
plt.show()
```
